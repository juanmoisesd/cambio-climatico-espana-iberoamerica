Análisis de emisiones de CO₂, temperatura media y eventos climáticos extremos en la región iberoamericana.

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import numpy as np

plt.rcParams['figure.dpi'] = 120
sns.set_theme(style='whitegrid')
print('Librerías cargadas ✓')
````

## 1. Emisiones de CO₂ per Cápita (toneladas, 2000–2024)

````python
data_co2 = {
    'Pais': ['España','México','Brasil','Argentina','Colombia','Chile','Perú'],
    '2000': [8.8, 3.7, 2.3, 3.9, 1.6, 4.3, 1.4],
    '2005': [9.1, 4.2, 2.5, 4.5, 1.9, 4.7, 1.7],
    '2010': [7.8, 4.0, 2.7, 4.8, 2.0, 4.6, 1.9],
    '2015': [6.4, 3.8, 2.6, 4.5, 2.1, 4.8, 2.1],
    '2020': [5.1, 3.2, 2.2, 4.0, 1.9, 4.5, 2.0],
    '2024': [4.8, 3.5, 2.4, 4.3, 2.2, 4.9, 2.3]
}
df_co2 = pd.DataFrame(data_co2).set_index('Pais')

data_temp = {
    'Año': list(range(2000, 2025)),
    'España': [14.5+i*0.035+np.random.normal(0,0.15) for i in range(25)],
    'México': [17.2+i*0.042+np.random.normal(0,0.12) for i in range(25)],
    'Brasil': [25.1+i*0.038+np.random.normal(0,0.1) for i in range(25)]
}
np.random.seed(42)
df_temp = pd.DataFrame(data_temp).set_index('Año')
print('Datos cargados:', df_co2.shape, df_temp.shape)
````

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

ax1 = axes[0]
years = ['2000','2005','2010','2015','2020','2024']
for pais in df_co2.index:
    ax1.plot(years, df_co2.loc[pais, years], marker='s', linewidth=2.5, label=pais)
ax1.set_title('Emisiones CO₂ per Cápita (t/habitante)\n2000–2024', fontsize=13, fontweight='bold')
ax1.set_xlabel('Año')
ax1.set_ylabel('Toneladas CO₂ per cápita')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2 = axes[1]
for col in df_temp.columns:
    ax2.plot(df_temp.index, df_temp[col], linewidth=2, label=col, alpha=0.8)
    z = np.polyfit(df_temp.index, df_temp[col], 1)
    p = np.poly1d(z)
    ax2.plot(df_temp.index, p(df_temp.index), '--', alpha=0.5)
ax2.set_title('Temperatura Media Anual (°C)\ncon tendencia lineal', fontsize=13, fontweight='bold')
ax2.set_xlabel('Año')
ax2.set_ylabel('Temperatura (°C)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('cambio_climatico_iberoamerica.png', dpi=150, bbox_inches='tight')
plt.show()
````
